# Recommendation System with Movielens Dataset

```python
#importing libraries
import numpy as np
import pandas as pd
```

```python
movie_data = pd.read_csv('datasets/ratings.csv')   #loads the dataset 'ratings.csv' from the datasets folder
movie_data.head(10)  #display the first ten rows of the ratings dataset
```

```python
movies = pd.read_csv('datasets/movies.csv')   #loads the dataset 'movies.csv' from the datasets folder
movies.head(10)  #display the first ten rows of the movies dataset
```

```python
tags = pd.read_csv('datasets/tags.csv') #loads the dataset 'movies.csv' from the datasets folder
tags = tags[["movieId","tag"]]
tags.head(10) #display the first ten rows of the movies dataset
```

```python
movie_data = movie_data.merge(movies,on='movieId',how='left') #left join rankings.csv and movies.cvs on movieId
movie_data = movie_data.merge(tags,on='movieId',how='left') #left join the result from above with tags.csv on movieId
movie_data.head(10)
```

```python
rating = pd.DataFrame(movie_data.groupby('title')['rating'].mean())  #group by movie_data by the mean of the rating
rating.head(10)
```

```python
rating['total rating'] = pd.DataFrame(movie_data.groupby('title')['rating'].count())  # adding a column for the number of rating per title
rating.head(10)
```

```python
movie_user = movie_data.pivot_table(index='userId', columns='title', values='rating') #correlation between users and rating given by users to a particular movie
movie_user.head(10)
```

```python
correlation = movie_user.corrwith(movie_user['Iron Man (2008)'])
correlation.head(10)
```

```python
recommendation = pd.DataFrame(correlation, columns=['correlation'])
recommendation.dropna(inplace=True)
recommendation = recommendation.join(rating['total rating'])
recommendation.head()
```

```python
recc = recommendation[recommendation['total rating']>150].sort_values('correlation',ascending=False)
recc = recc.merge(movies,on='title',how='left')
recc.head(10)
```
